# Problem 1
Write a function that checks whether a given list of numbers is sorted in either ascending or descending order.

```
# Sample input list
number = [2, 3, 4, 5]

# Define the function
def problem_1(nums):
    # Initialize flags to check order
    is_ascending = True
    is_descending = True

    # Check for ascending or descending order
    for i in range(len(nums) - 1):
        if nums[i] > nums[i + 1]:
            is_descending = False
        if nums[i] < nums[i + 1]:
            is_ascending = False

    # Return True if the list is either ascending or descending
    return is_ascending or is_descending

# Test the function
print(problem_1(number))  # Expected output: True
```

# Problem 2
Write a function that takes a credit card number and transforms it into a string where all digits except the last four are replaced with asterisks.

```
# Sample input
credit_card = "1234567812345678"

# Define the function
def mask_credit_card(card_number):
    # Replace all digits except the last four with asterisks
    masked = "*" * (len(card_number) - 4) + card_number[-4:]
    return masked

# Test the function
print(mask_credit_card(credit_card))  # Expected output: ************5678
```
