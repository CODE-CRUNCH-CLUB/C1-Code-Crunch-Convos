## 2. Load the data

We try seaborn first. If that fails for any reason, we read a tiny inline
CSV with eight rows so you still have a real DataFrame to play with.

```python
try:
    import seaborn as sns
    df = sns.load_dataset("tips")
    print("Loaded the seaborn `tips` dataset.")
except Exception as exc:
    print("seaborn not available, falling back to inline CSV:", exc)
    inline_csv = (
        "total_bill,tip,sex,smoker,day,time,size\n"
        "16.99,1.01,Female,No,Sun,Dinner,2\n"
        "10.34,1.66,Male,No,Sun,Dinner,3\n"
        "21.01,3.50,Male,No,Sun,Dinner,3\n"
        "23.68,3.31,Male,No,Sun,Dinner,2\n"
        "24.59,3.61,Female,No,Sun,Dinner,4\n"
        "25.29,4.71,Male,No,Sun,Dinner,4\n"
        "8.77,2.00,Male,No,Sun,Dinner,2\n"
        "26.88,3.12,Male,No,Sun,Dinner,4\n"
    )
    df = pd.read_csv(io.StringIO(inline_csv))

type(df)
```

## 3. First look

## 4. Shape and dtypes

`shape` is a tuple `(rows, columns)`. `dtypes` shows the type of each
column. `object` typically means strings.

```python
print("shape:", df.shape)
df.dtypes
```

## 5. `info()` — the kitchen-sink summary

`info()` shows row count, columns, non-null counts, dtypes, and memory.

```python
df.info()
```

## 6. `describe()` — summary statistics

By default `describe()` only summarizes numeric columns. Pass
`include="all"` to also see counts and most-frequent values for string
columns.

```python
df.describe()
```

```python
df.describe(include="all")
```

## 7. `value_counts` for categorical columns

How many bills fell on each day?

```python
df["day"].value_counts()
```
